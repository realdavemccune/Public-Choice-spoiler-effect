```python
def IRV(frame):
    num_cands_remaining=len(frame.iloc[0]['Cand List'])
    num_cands=num_cands_remaining
    cand_dict={ }
    keys=range(num_cands)
    values=[]
    
    for k in range(len(frame.iloc[0]['Cand List'])):
        values.append(frame.iloc[0]['Cand List'][k])
    
    for i in keys:
        cand_dict[i]=values[i]
        
    cand_list=[]
    for item in cand_dict:
        
        cand_list.append(cand_dict[item])
    #print(cand_dict)
    #print(cand_list)
        
    num_cands_remaining=num_cands
    while num_cands_remaining>1:
        
        scores=[0 for i in range(num_cands)]
        for k in range(len(frame)):
            if frame.iloc[k]['ballot']!='':
                l=cand_list.index(frame.iloc[k]['ballot'][0])
                scores[l]+=frame.iloc[k]['Count']
        #print(scores)
        remaining_vote=0
        for k in range(len(frame)):
            if frame.iloc[k]['ballot']!='':
                remaining_vote+=frame.iloc[k]['Count']
        for k in range(len(scores)):
            if scores[k]>remaining_vote/2:
                return [str(cand_dict[k])]
        
        min_score=min(i for i in scores if i>0)
        positions_of_lowest_score=[i for i,v in enumerate(scores) if v==min_score]
        #print(positions_of_lowest_score)
        
        #Tie at the top, record the multiple winners
        if num_cands_remaining==len(positions_of_lowest_score):
            return [cand_dict[i] for i in positions_of_lowest_score]
            break
        num_cands_remaining-=len(positions_of_lowest_score)
        
        for k in range(len(frame)):
            for l in positions_of_lowest_score:
                frame.at[k,'ballot']=frame.at[k,'ballot'].replace(cand_dict[l],'')
        
        frame=pd.DataFrame(frame.groupby(['ballot'],as_index=False)['Count'].sum())
        
            
    
    
    for k in range(len(frame)):
        if frame.iloc[k]['ballot']!='':
            return [frame.iloc[k]['ballot'][0]]
        
def plurality(frame):
    num_cands_remaining=len(frame.iloc[0]['Cand List'])
    num_cands=num_cands_remaining
    cand_dict={ }
    keys=range(num_cands)
    values=[]
    
    for k in range(len(frame.iloc[0]['Cand List'])):
        values.append(frame.iloc[0]['Cand List'][k])
    
    for i in keys:
        cand_dict[i]=values[i]
        
    cand_list=[]
    for item in cand_dict:
        
        cand_list.append(cand_dict[item])
        
    num_cands_remaining=num_cands
   
        
    scores=[0 for i in range(num_cands)]
    for k in range(len(frame)):
        if frame.iloc[k]['ballot']!='':
            l=cand_list.index(frame.iloc[k]['ballot'][0])
            scores[l]+=frame.iloc[k]['Count']
    
    max_score=max(scores)

    positions_of_max_score=[i for i,v in enumerate(scores) if v==max_score]
        #print(positions_of_lowest_score)
    return [cand_dict[i] for i in positions_of_max_score]    


def powerset(seq):
    """
    Returns all the subsets of this set. This is a generator.
    """
    if len(seq) <= 1:
        yield seq
        yield []
    else:
        for item in powerset(seq[1:]):
            yield [seq[0]]+item
            yield item

def Cond_loser(frame):
    num_cands=len(frame.iloc[0]['Cand List'])
    cand_dict={ }
    keys=range(num_cands)
    values=[]
    
    for k in range(len(frame.iloc[0]['Cand List'])):
        values.append(frame.iloc[0]['Cand List'][k])
    
    for i in keys:
        cand_dict[i]=values[i]
        
    cand_list=[]
    for item in cand_dict:
        cand_list.append(cand_dict[item])
    Cond_scores=[0 for i in range(len(cand_dict))]
    #Get all pairings
    possible_pairings=[]
    for i in range(len(cand_dict)):
        for j in range(len(cand_dict)):
            if i<j:
                possible_pairings.append([cand_dict[i],cand_dict[j]])

    for pairing in possible_pairings:
        #print(pairing)
        pairing_scores=[0,0]
        Cand1=pairing[0]
        Cand2=pairing[1]
        for k in range(len(frame)):
            for l in range(len(frame.iloc[k]['ballot'])):
                if frame.iloc[k]['ballot'][l]==Cand1:
                    pairing_scores[0]+=frame.iloc[k]['Count']
                    #print(Cand1, cands.index(Cand1),pairing_scores)
                    break
                if frame.iloc[k]['ballot'][l]==Cand2:
                    pairing_scores[1]+=frame.iloc[k]['Count']
                    #print(Cand2, cands.index(Cand2),pairing_scores)
                    break
        #print(pairing_scores)
        if pairing_scores[0]>pairing_scores[1]:
            Cond_scores[cand_list.index(Cand1)]+=1
        if pairing_scores[0]<pairing_scores[1]:
            Cond_scores[cand_list.index(Cand2)]+=1
        
    
    if 0 not in Cond_scores:
        return False
    else:
        return cand_dict[Cond_scores.index(0)]
    
def Cond_winner(frame):
    num_cands=len(frame.iloc[0]['Cand List'])
    cand_dict={ }
    keys=range(num_cands)
    values=[]
    
    for k in range(len(frame.iloc[0]['Cand List'])):
        values.append(frame.iloc[0]['Cand List'][k])
    
    for i in keys:
        cand_dict[i]=values[i]
        
    cand_list=[]
    for item in cand_dict:
        cand_list.append(cand_dict[item])
    Cond_scores=[0 for i in range(len(cand_dict))]
    #Get all pairings
    possible_pairings=[]
    for i in range(len(cand_dict)):
        for j in range(len(cand_dict)):
            if i<j:
                possible_pairings.append([cand_dict[i],cand_dict[j]])

    for pairing in possible_pairings:
        print(pairing)
        pairing_scores=[0,0]
        Cand1=pairing[0]
        Cand2=pairing[1]
        for k in range(len(frame)):
            for l in range(len(frame.iloc[k]['ballot'])):
                if frame.iloc[k]['ballot'][l]==Cand1:
                    pairing_scores[0]+=frame.iloc[k]['Count']
                    #print(Cand1, cands.index(Cand1),pairing_scores)
                    break
                if frame.iloc[k]['ballot'][l]==Cand2:
                    pairing_scores[1]+=frame.iloc[k]['Count']
                    #print(Cand2, cands.index(Cand2),pairing_scores)
                    break
        print(pairing_scores)
        if pairing_scores[0]>pairing_scores[1]:
            Cond_scores[cand_list.index(Cand1)]+=1
        if pairing_scores[0]<pairing_scores[1]:
            Cond_scores[cand_list.index(Cand2)]+=1
    
    
    if (num_cands-1) not in Cond_scores:
        return False
    else:
        return cand_dict[Cond_scores.index(num_cands-1)]





            
def clones(frame,cand):
    cand_set=[char for char in frame.iloc[0]['Cand List']]
    #cand_set=list(frame.iloc[0]['Cand List'])
    cand_set.remove(cand)
    clone_set=[]
    n=len(frame.iloc[0]['Cand List'])
    
    for cand in cand_set:
        clone=True
        
        for k in range(len(frame)):
            if abs(frame.iloc[k]['ballot'].index(cand)-frame.iloc[k]['ballot'].index(str(l)))>1:
                clone=False
                break
        if clone:
            clone_set.append(cand)
    return clone_set

def almost_clones(frame, l, p):
    cand_set=list(frame.iloc[0]['ballot'])
    cand_set.remove(str(l))
    clone_set=[]
    n=len(frame.iloc[0]['ballot'])
    for cand in cand_set:
        clone=True
        count=0
        for k in range(len(frame)):
            if abs(frame.iloc[k]['ballot'].index(cand)-frame.iloc[k]['ballot'].index(str(l)))==1:
                count+=frame.iloc[k]['Count']
        if count/sum(frame['Count'])>p:
            clone_set.append(cand)
    return clone_set
```

```python
filename='APA 2020 BoD race2'
data_original = pd.read_csv("Desktop/ranked choice datasets cleaned for Public Choice simulations/cleaned versions final/"+filename+".csv")
#data.to_csv('Desktop/Dem Borough President Bronx.csv')
electorate_size=sum(data_original['Count'])
n=len(data_original.iloc[0]['Cand List'])

if electorate_size>10000:
    electorate_size=10000

    
CL_plurality_spoiler=0
CL_IRV_spoiler=0
plur_spoiler=0
IRV_spoiler=0
plur_spoiler_set=0
IRV_spoiler_set=0

pos_list=[i for i in range(len(data_original))]
prob_dist=[data_original.iloc[i]['Count']/sum(data_original['Count']) for i in range(len(data_original))]

a=10000


while a>0:
    column_names=['ballot','Count']
    data=pd.DataFrame(columns = column_names)
    data['ballot']=data_original['ballot']
    data['Count']=[0 for i in range(len(data_original))]
    data.loc[data.index[0],'Cand List']=data_original.iloc[0]['Cand List']
    
    
    
    for t in range(electorate_size):
        x=int(random.choices(pos_list, prob_dist)[0])
        data.at[x,'Count']+=1                                 
   
    
    #Check for majority candidate
    scores=[0 for i in range(n)]
    cand_list=[]
    for char in data.iloc[0]['Cand List']:
        cand_list.append(char)
    for k in range(len(data)):
        for cand in cand_list:
            if data.iloc[k]['ballot'][0]==cand: 
                    scores[cand_list.index(cand)]+=data.iloc[k]['Count']
    
    
    
    if max(scores)>electorate_size/2:
        continue
    data_copy=data.copy(deep=True)    
    IRV_winners=IRV(data_copy)
    data_copy=data.copy(deep=True)  
    plurality_winners=plurality(data_copy)
    CL=Cond_loser(data_copy)
    
    
    cands=[i for i in data_original.iloc[0]['Cand List']]
    r=[x for x in powerset(cands)]
    
    #Check for plurality spoiler candidates and sets
    
    #First check for spoiler candidates
    plurality_spoiler=False
    plurality_spoiler_set=False
    
    for cand in cands:
        if cand not in plurality_winners:
            data_copy=data.copy(deep=True)
            for b in range(len(data_copy)):
                data_copy.at[b,'ballot']=data_copy.at[b,'ballot'].replace(cand,'')
            data_copy.at[0,'Cand List']=data_copy.at[0,'Cand List'].replace(cand,'')
            #print(data_copy)
            new_plur_winners=plurality(data_copy)
            #print(new_plur_winners)
            if plurality_winners!=new_plur_winners:
                plurality_spoiler=True
                plurality_spoiler_set=True
                if cand==CL:
                    CL_plurality_spoiler+=1
                    
    
    if plurality_spoiler==False:
        for subset in r:
            if len(subset)>=2 and set(plurality_winners).isdisjoint(set(subset)):
                #print(l)
                data_copy=data.copy(deep=True)
                for b in range(len(data_copy)):
                    for cand in subset:
                        data_copy.at[b,'ballot']=data_copy.at[b,'ballot'].replace(cand,'')
                for cand in subset:
                    data_copy.at[0,'Cand List']=data_copy.at[0,'Cand List'].replace(cand,'')

                new_plur_winners=plurality(data_copy)
                #print(new_plur_winners)
                if plurality_winners!=new_plur_winners:
                    plurality_spoiler_set=True
                
                
    if plurality_spoiler:
        plur_spoiler+=1
    if plurality_spoiler_set:
        plur_spoiler_set+=1
    
    IRV_spoiler_exists=False
    IRV_spoiler_set_exists=False
    #Check for IRV spoiler candidates and sets
    #First, check for spoiler cands
    for cand in cands:
        if cand not in IRV_winners:
            data_copy=data.copy(deep=True)
            for b in range(len(data_copy)):
                data_copy.at[b,'ballot']=data_copy.at[b,'ballot'].replace(cand,'')
            data_copy.at[0,'Cand List']=data_copy.at[0,'Cand List'].replace(cand,'')
            new_IRV_winners=IRV(data_copy)
            if IRV_winners!=new_IRV_winners:
                IRV_spoiler_exists=True
                IRV_spoiler_set_exists=True
                
                if cand==CL:    
                    CL_IRV_spoiler+=1
    
    if IRV_spoiler_exists==False:
        for subset in r:
            if len(subset)>=2 and set(IRV_winners).isdisjoint(set(subset)):
                #print(l)
                data_copy=data.copy(deep=True)
                for b in range(len(data_copy)):
                    for cand in subset:
                        data_copy.at[b,'ballot']=data_copy.at[b,'ballot'].replace(cand,'')
                for cand in subset:
                    data_copy.at[0,'Cand List']=data_copy.at[0,'Cand List'].replace(cand,'')

                new_IRV_winners=IRV(data_copy)
                #print(new_plur_winners)
                if IRV_winners!=new_IRV_winners:
                    if len(subset)==1:
                        IRV_spoiler_exists=True
                    IRV_spoiler_set_exists=True
                
    if IRV_spoiler_exists:
        IRV_spoiler+=1
    if IRV_spoiler_set_exists:
        IRV_spoiler_set+=1
    
    
    a-=1

print(filename)
print("Num elections with a plurality spoiler cand: " ,plur_spoiler)
print("Num elections with a plurality spoiler set: " ,plur_spoiler_set)
print("Number of elections in which CL is a plurality spoiler ", CL_plurality_spoiler)
print("Num elections with an IRV spoiler cand: ", IRV_spoiler)
print("Num elections with an IRV spoiler set: ", IRV_spoiler_set)
print("Number of elections in which CL is an IRV spoiler ", CL_IRV_spoiler)

            

```

```python
def IRV2(frame):
    num_cands_remaining=len(frame.iloc[0]['Cand List'])
    num_cands=num_cands_remaining
    cand_dict={ }
    keys=range(num_cands)
    values=[]
    
    for k in range(len(frame.iloc[0]['Cand List'])):
        values.append(frame.iloc[0]['Cand List'][k])
    
    for i in keys:
        cand_dict[i]=values[i]
        
    cand_list=[]
    for item in cand_dict:
        
        cand_list.append(cand_dict[item])
    #print(cand_dict)
    #print(cand_list)
    print(cand_dict)  
    num_cands_remaining=num_cands
    while num_cands_remaining>1:
        
        scores=[0 for i in range(num_cands)]
        for k in range(len(frame)):
            if frame.iloc[k]['ballot']!='':
                l=cand_list.index(frame.iloc[k]['ballot'][0])
                scores[l]+=frame.iloc[k]['Count']
        print(scores)
        remaining_vote=0
        for k in range(len(frame)):
            if frame.iloc[k]['ballot']!='':
                remaining_vote+=frame.iloc[k]['Count']
        for k in range(len(scores)):
            if scores[k]>remaining_vote/2:
                return [str(cand_dict[k])]
        
        min_score=min(i for i in scores if i>0)
        positions_of_lowest_score=[i for i,v in enumerate(scores) if v==min_score]
        #print(positions_of_lowest_score)
        
        #Tie at the top, record the multiple winners
        if num_cands_remaining==len(positions_of_lowest_score):
            return [cand_dict[i] for i in positions_of_lowest_score]
            break
        num_cands_remaining-=len(positions_of_lowest_score)
        
        for k in range(len(frame)):
            for l in positions_of_lowest_score:
                frame.at[k,'ballot']=frame.at[k,'ballot'].replace(cand_dict[l],'')
        
        frame=pd.DataFrame(frame.groupby(['ballot'],as_index=False)['Count'].sum())
        
            
    
    
    for k in range(len(frame)):
        if frame.iloc[k]['ballot']!='':
            return [frame.iloc[k]['ballot'][0]]

filename='2008_Pierce_County_Executive'
data_original = pd.read_csv("Desktop/ranked choice datasets cleaned for Public Choice simulations/cleaned versions final/"+filename+".csv")
#data.to_csv('Desktop/Dem Borough President Bronx.csv')
IRV2(data_original)
```
